```python
# Task 2: Mark the success/failed launches for each site on the map


# Create map centered on all launch sites
launch_map = folium.Map(
    location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()],
    zoom_start=5
)

# Loop through each launch record
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    site_name = row['Launch Site']
    outcome = row['class']  # 1 = success, 0 = failure

    # Choose marker color
    color = "green" if outcome == 1 else "red"
    outcome_text = "Success" if outcome == 1 else "Failed"

    # Add marker to map
    folium.Marker(
        location=coordinate,
        popup=f"{site_name}<br>{outcome_text}",
        icon=folium.Icon(color=color, icon="rocket", prefix="fa")
    ).add_to(launch_map)

launch_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

def haversine_distance(lat1, lon1, lat2, lon2):
    # Earth radius in km
    R = 6371.0
    
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c   # distance in kilometers
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python

poi_data = {
    'Railway': {'lat': 28.5621, 'lon': -80.5770},
    'Highway': {'lat': 28.5610, 'lon': -80.5790},
    'Coastline': {'lat': 28.5630, 'lon': -80.5800}
}


```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2-lat1, lon2-lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c
```

```python
import folium

m = folium.Map(location=[launch_lat, launch_lon], zoom_start=13)

# Launch site marker
folium.Marker([launch_lat, launch_lon], popup="Launch Site", icon=folium.Icon(color='red')).add_to(m)

# Add POIs and lines
for poi, coords in poi_data.items():
    dist = haversine(launch_lat, launch_lon, coords['lat'], coords['lon'])
    
    # POI marker with distance in popup
    folium.Marker(
        [coords['lat'], coords['lon']],
        popup=f"{poi}: {dist:.2f} km",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)
    
    # Line between launch site and POI
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], [coords['lat'], coords['lon']]],
        color='green', weight=2, opacity=0.7
    ).add_to(m)

m
```
